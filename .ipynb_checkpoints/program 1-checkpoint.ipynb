# Write a program for word analysis


Word Tokenization

```python
from nltk.tokenize import word_tokenize
words = "A sentence is a combination of words put together to convey an idea, a fact, a question, a thought, a request or a command. Does that mean that you can assemble words in any order? Will it still be a sentence? No! Learn what a sentence is and how they are formed in this article. Furthermore, go through the components of a sentence, the types of sentences, and the given examples to understand how they are structured."
```

```python
tokenizedword = word_tokenize(words)
```

```python
print(tokenizedword)
```

Frequency Distribution

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokenizedword)
```

```python
print(fdist.most_common(4))
```

Frequency Distribution on Graph

```python
import matplotlib.pyplot as plt
fdist.plot(30 , cumulative=False)
plt.show()
```

Stopwords

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stopwords = stopwords.words('english')
```

Filtering tokenized words

```python
filtered_tokens = []
for w in tokenizedword:
    if w not in stopwords:
        filtered_tokens.append(w)
        
print("Tokenized words:" , tokenizedword)
print(" ")
print("Filtered words:" , filtered_tokens)
```

Removing punctuations from filtered tokens

```python
import string
punctuations = list(string.punctuation)
filtered_tokens2 =[]

for i in filtered_tokens:
    if i not in punctuations:
        filtered_tokens2.append(i)
print("filtered tokens after removing punctuations:" , filtered_tokens2)
```

```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

nltk.download("stopwords")
```

```python
text = "essay, an analytic, interpretative, or critical literary composition usually much shorter and less systematic and formal than a dissertation or thesis and usually dealing with its subject from a limited and often personal point of view. Category: Arts & Culture."
```

```python
words = word_tokenize(text)
```

```python
print(words)
```

```python
stop_words = set(stopwords.words("english"))
```

```python
filtered_words = [word.lower() for word in words if word.isalnum() and word.lower() not in stop_words]
```

```python
wordcloud = WordCloud(width=800 , height=400 ,background_color="white" ).generate(" ".join(filtered_words))
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud , interpolation="bilinear")
plt.axis("off")
plt.show()
```
